```python
import os
import shutil

path = input("Enter Path: ")
files = os.listdir(path)

for file in files:
    filename, extension = os.path.splitext(file)
    extension = extension[1:]

    if os.path.exists(path+'/'+extension):
        shutil.move(path+'/'+file, path+'/'+extension+'/'+file)
    else:
        os.makedirs(path+'/'+extension)
        shutil.move(path+'/'+file, path+'/'+extension+'/'+file)

        
```

```python
import os
import shutil
import tkinter as tk
from tkinter import filedialog, messagebox

# Function to organize files
def organize_files():
    path = filedialog.askdirectory(title="Select Folder to Organize")
    if not path:
        messagebox.showwarning("Warning", "No folder selected!")
        toggle_button.config(text="Start Organizing", state="normal")
        return

    files = os.listdir(path)
    for file in files:
        filename, extension = os.path.splitext(file)
        extension = extension[1:]

        if extension == "":
            continue

        folder_path = os.path.join(path, extension)
        source_path = os.path.join(path, file)

        if not os.path.exists(folder_path):
            os.makedirs(folder_path)

        shutil.move(source_path, os.path.join(folder_path, file))

    messagebox.showinfo("Success", "Files organized successfully!")
    toggle_button.config(text="Start Organizing", state="normal")

# Toggle action handler
def toggle_action():
    toggle_button.config(text="Organizing...", state="disabled")
    root.after(100, organize_files)

# Create main window
root = tk.Tk()
root.title("Automatic File Organizer")
root.geometry("420x270")
root.configure(bg="#f2f7f5")  # Light pastel background
root.resizable(False, False)

# Styling: Fonts and spacing
header_font = ("Segoe UI", 16, "bold")
button_font = ("Segoe UI", 11)

# Heading Label
label = tk.Label(root, text="Automatic File Organizer", font=header_font,
                 bg="#f2f7f5", fg="#2e2e2e", pady=20)
label.pack()

# Instruction label
instruction = tk.Label(root, text="Select a folder to automatically organize files by type.",
                       font=("Segoe UI", 10), bg="#f2f7f5", fg="#4b4b4b")
instruction.pack(pady=5)

# Toggle Button
toggle_button = tk.Button(root, text="Start Organizing", command=toggle_action,
                          bg="#4da6ff", fg="white", font=button_font,
                          activebackground="#3399ff", activeforeground="white",
                          padx=15, pady=8, bd=0, relief="ridge", cursor="hand2")
toggle_button.pack(pady=20)

# Footer Label
footer = tk.Label(root, text="© 2025 Nitesh Solanki", font=("Segoe UI", 9),
                  bg="#f2f7f5", fg="gray")
footer.pack(side="bottom", pady=10)

# Start GUI
root.mainloop()
```
